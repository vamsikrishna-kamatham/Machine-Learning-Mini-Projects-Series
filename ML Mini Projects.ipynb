<h2 style="color:green"> 3.0. Math behind ML Concepts </h2><a id='3.0'></a>

<h2 style="color:greY"> Mean Square Error Cost Function, Gradient Descent, and Learning Rate</h3>

Mean square error cost function and gradient descent are important concepts in machine learning. Machine learning uses Mean Square Error (MSE) cost function and gradient descent to find the best fit line for a given set of input and output data. The cost function calculates the error between the predicted and actual data points, while gradient descent adjusts the parameters of the prediction function to minimize the cost function. The learning rate determines the step size for each iteration of gradient descent, with a high learning rate causing faster convergence but potentially overshooting the optimal solution, and a low learning rate causing slower convergence but potentially resulting in a more accurate solution.

<h2 style="color:green"> 5.0. Dummy Variables & One Hot Encoding </h3><a id='5.0'></a>

Machine learning models are good at handling numeric data, but text data requires encoding. Label encoding or integer encoding can be used to convert text data into numeric form.However, label encoding assumes order relationships that may not exist in nominal variables.Nominal variables are a type of categorical variable where categories have no order relationship with each other.Ordinal variables are another type of categorical variable where categories have a numerical ordering relationship.One-hot encoding is a technique used to solve this problem, where a new column is created for each category, and binary values of 0 or 1 are assigned to each column. The new columns created through one-hot encoding are called dummy variables.

<h2 style="color:green">7.0. Logistic Regression (Binary Classification)</h2><a id='7.0'></a>

<img height=200 width=600 src="22.png" />

Logistic regression is a technique used to solve classification problems, which involve predicting categorical outcomes rather than continuous ones. In contrast to linear regression, where the predicted value can be any number, in logistic regression, the predicted value is a probability that the input belongs to a particular category. Logistic regression works by transforming the input using a logistic function, which maps the input onto a probability between 0 and 1. The logistic function is a type of sigmoid function, which is an S-shaped curve that is widely used in many applications.

Types of classification: There are two types of classification problems: binary classification and multi-class classification. In binary classification, the outcome is a simple yes or no, while in multi-class classification, there are more than two categories. For example, predicting whether a customer will buy insurance or not is a binary classification problem, while predicting which political party a person will vote for is a multi-class classification problem.

As an example, let's say a data scientist is tasked with predicting whether potential customers will buy a life insurance product based on their age. By plotting the data, the scientist can see that age is related to the likelihood of purchasing insurance. The scientist may be tempted to use linear regression to predict the outcome, but this would not be appropriate since the outcome is categorical. Instead, logistic regression can be used to predict the probability of a customer buying the insurance product based on their age. A threshold can then be set, such as 0.5, and any probability greater than the threshold can be classified as a yes, while any probability less than the threshold can be classified as a no.

<h2 style="color:green">9.0. Decision Tree</h2><a id='9.0'></a>

<img height=200 width=600 src="23.png"/>

Decision trees are an algorithm used for classification problems in machine learning. They split data into subsets based on decision rules, with the goal of creating a tree that can accurately classify new data. The algorithm selects features with the most information gain to split the data, and this process is repeated recursively until a stopping criterion is met. Decision trees can be used for various classification problems, and the order in which features are split affects the algorithm's performance. Features can be ordered based on information gain and entropy, with low entropy indicating low randomness and high information gain indicating more accurate predictions. Decision trees are more suitable for complex datasets compared to logistic regression, which is best for simpler datasets with a single decision boundary.

In practice, decision trees can be used for a variety of classification problems. For example, predicting a person's salary based on the company, job title, and degree they have can be solved using a decision tree. The algorithm would split the data based on the company first, then further split the data based on job title and degree. The order in which the features are split would be determined based on the information gain provided by each feature.

<img height=200 width=600 src="25.png"/>

SVM is an algorithm used for classification by drawing a boundary that separates different classes in the data space. To determine the decision boundary, nearby data points called support vectors are used. The best boundary is the one with a higher margin, which classifies the two groups better. In 2D, the boundary is a line, in 3D it is a plane, and in n-dimensional space, it is a hyperplane. SVM can be used for classification problems in any dimension. Gamma and regularization are technical terms associated with SVM. Gamma refers to how far the influence of a single training example reaches, and regularization prevents overfitting by adding a penalty term to the loss function.

<h2 style="color:green">11.0. Random Forest</h2><a id='11.0'></a>

<img height=200 width=800 src="24.png"/>

Random forest algorithm is a popular machine learning technique used in regression and classification tasks. The name "random forest" comes from the fact that the algorithm builds multiple decision trees based on random subsets of the data and combines their predictions to make a final decision. Each decision tree is built on a different subset of the data and with a different set of features. The algorithm then takes the majority vote of all the decision trees to make a final prediction.The basic crux behind the random forest algorithm is to take a majority decision from multiple decision trees.

To build a random forest, the data set is divided into batches of random data sets, and a decision tree is built for each batch. Once trained, the algorithm can be used to predict the outcome for new data by using the majority vote of all the decision trees in the forest.

<h2 style="color:green">12.0. K Fold Cross Validation</h2><a id='12.0'></a>

<img height=200 width=800 src="32.png"/>

The topic discusses various techniques to evaluate a machine learning model's performance. One of the most common techniques used is cross-validation. Cross-validation allows the model's performance to be evaluated by training the model on a dataset and testing it on a different dataset. This helps to avoid overfitting, where the model performs well on the training dataset but not on the testing dataset.

The topic discusses three ways to train a model:  
Train the model on all available data and test it on the same dataset.  
Split the available dataset into training and testing datasets.  
Use k-fold cross-validation, where the dataset is divided into k equal parts, and the model is trained and tested k times using different combinations of training and testing datasets.  

The topic highlights the limitations of the first two techniques and suggests that k-fold cross-validation is a more robust technique for evaluating a model's performance. The k-fold cross-validation technique avoids the problem of overfitting and ensures that the model's performance is measured using different combinations of training and testing datasets.

<h2 style="color:green">13.0. K Means Clustering Algorithm</h2> <a id='13.0'></a>

<img height=200 width=400 src="33.png"/>

Supervised learning and unsupervised learning are two main categories of machine learning algorithms. In supervised learning, the data set contains a class label or a target variable that the algorithm learns to predict. In contrast, unsupervised learning involves a data set with only features, and the algorithm tries to identify the underlying structure in the data or find clusters. K-means clustering is a popular unsupervised learning algorithm that helps to identify clusters in a data set. In k-means, the algorithm tries to group data points into k clusters based on their similarity. The algorithm starts with randomly placed centroids for each cluster and then adjusts them iteratively to find better cluster centers. The process continues until the algorithm converges to a stable solution. Elbow method is a technique used to determine optimal number of k, we will review that method as well.

Naive Bayes is a classification algorithm used in machine learning. It makes use of Bayes' theorem, which is a powerful tool to find conditional probability. To understand Naive Bayes, it's important to have some basic understanding of probability theory, including concepts such as conditional probability and Bayes' theorem. Bayes' theorem is a powerful theorem that allows us to find conditional probabilities when we know some individual probabilities but not others.

In the context of machine learning, Naive Bayes can be used to classify data based on the probability of certain features. For example, in the case of the Titanic crash, we might use Naive Bayes to calculate the probability of survival based on features such as age, gender, fare, and cabin. While Naive Bayes assumes that these features are independent of each other, in reality, some features may be dependent. <h2 style="color:green">16.0. Hyper parameter Tuning (GridSearchCV)</h2><a id='16.0'></a>

The topic is about how to choose the best model for a given machine learning problem and how to do hyperparameter tuning. The example used in the topic is classifying the SKLearn Iris Flower dataset using the SVM model. The traditional approach to solve this problem is to use the train-test split method to split the dataset into training and test datasets. However, this method is not reliable because the score is changing based on the samples. For this reason, the K-fold cross-validation method is used. This method divides the data samples into n number of folds, takes one fold as the test set in each iteration, and finds the score for that iteration. The individual scores from each iteration are averaged to find the optimal value for the hyperparameters(cross-validation with different values of kernels and C to determine the optimal value of these parameters). The manual and repetitive method of finding the optimal values for the hyperparameters using the K-fold cross-validation method is shown, and a more efficient method of using a for loop to do the same thing is suggested.

<h3>Approach 2: Use K Fold Cross validation</h3>

Manually try suppling models with different parameters to cross_val_score function with 5 fold cross validation

````python
# Approach 2: use K-fold cross validation to evaluate model performance with different hyperparameters
# manually supply models with different hyperparameters to cross_val_score function with 5-fold cross validation
cross_val_score(svm.SVC(kernel='linear',C=10,gamma='auto'),iris.data, iris.target, cv=5)
````

The process of manually testing different model parameters using the cross_val_score() function can be tedious and time-consuming. An alternative approach is to use a for loop to automate the process. This can save time and effort by testing different parameter values automatically.

````python
# use for loop to automate the process of testing different hyperparameters
kernels = ['rbf', 'linear']
C = [1,10,20]
avg_scores = {}
for kval in kernels:
    for cval in C:
        cv_scores = cross_val_score(svm.SVC(kernel=kval,C=cval,gamma='auto'),iris.data, iris.target, cv=5)
        avg_scores[kval + '_' + str(cval)] = np.average(cv_scores)
        
# display the average scores for each model configuration
avg_scores
````

1. Linear kernel with C=1

<h2 style="color:green">20.0. Bias vs Variance In Machine Learning </h2><a id='20.0'></a>      

<img height=200 width=400 src="36.png"/>

Bias variance trade off is a popular term in statistics. The concepts of bias and variance can be explained using a simple analogy of a t-shirt that can be overfit, underfit, or perfectly fit.where a perfectly fitting t-shirt represents a balanced model, an overfitting t-shirt represents an overfit model, and an underfitting t-shirt represents an underfit model.

In the context of machine learning, bias refers to the error due to the assumptions made by the model, while variance refers to the error due to the model's sensitivity to the training data. An overfit model is one that fits too closely to the training data, resulting in high training accuracy but poor generalization to new data. An underfit model is one that is too simple and fails to capture the patterns in the data, resulting in poor training accuracy and poor generalization to new data. A well-fit model is one that strikes a balance between bias and variance and achieves good performance on both the training and test data.

In summary an overfit model tries to fit exactly to the training samples, resulting in a low training error but a high test error. An underfit model has high training and high test errors, while a balanced model has low training and test errors. Finding the right balance between bias and variance is crucial for building a successful machine learning model.

<h2 style="color:green">21.0. Ensemble Learning - Bagging </h2><a id='21.0'></a>      

<img height=200 width=600 src="37.png"/>

Ensemble Learning: A technique used in machine learning to improve the accuracy of a model by combining predictions from multiple models. Ensemble learning can help address the high variance problem that can occur when a model is trained using all the data in a dataset, and there are two main techniques used in ensemble learning: bagging and boosting. Bagging works by creating smaller subsets of the original dataset using resampling with replacement, training a machine learning model on each subset, and combining the predictions from these models to produce a final output. An example of using bagging to classify whether a person should buy insurance or not using logistic regression is also provided.

In summary Ensemble learning is all about using multiple models to combine their prediction power to get better predictions that has low variance. Bagging and boosting are two popular techniques that allows us to tackle high variance issue.
